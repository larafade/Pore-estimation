```python
from PIL import Image
import os
from google.colab import drive

drive.mount('/content/drive/')
image_path = '/content/drive/My Drive/MASC 515/Project/RP_HL_LS_8_1_BottomCenter_4_150x_BW.jpg'
output_directory = '/content/drive/My Drive/MASC 515/Project/crop_final'


def crop_image(image_path, output_dir, crop_size=512):
    # Open the image
    image = Image.open(image_path)

    # Get the dimensions of the image
    width, height = image.size

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Iterate through the image and crop it into 512x512 pixel crops
    for y in range(0, height, crop_size):
        for x in range(0, width, crop_size):
            # Define the bounding box for cropping
            box = (x, y, x + crop_size, y + crop_size)

            # Crop the image
            cropped_img = image.crop(box)

            # If the cropped region is exactly 512x512, save it
            if cropped_img.size == (crop_size, crop_size):
                # Generate output filename
                filename = os.path.join(output_dir, f"cropped_image_{y // crop_size}_{x // crop_size}.jpg")
                # Save cropped image
                cropped_img.save(filename)


crop_image(image_path, output_directory)
```

```python
import os
from PIL import Image
from google.colab import drive

drive.mount('/content/drive/')
image_path = '/content/drive/My Drive/MASC 515/Project/RP_HL_SL_8_1_OutsideAlongBondline_1_150x_BW.jpg'
output_directory = '/content/drive/My Drive/MASC 515/Project/crop_final2/image1'


def crop_image(image_path, output_dir, crop_size=512, overlap=100):
    # Open the image
    image = Image.open(image_path)

    # Get the dimensions of the image
    width, height = image.size

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Calculate the number of crops in both dimensions
    num_crops_x = (width - crop_size) // (crop_size - overlap) + 1
    num_crops_y = (height - crop_size) // (crop_size - overlap) + 1

    # Iterate through the image and crop it
    for i in range(num_crops_y):
        for j in range(num_crops_x):
            # Calculate crop coordinates
            x = j * (crop_size - overlap)
            y = i * (crop_size - overlap)

            # Define the bounding box for cropping
            box = (x, y, x + crop_size, y + crop_size)

            # Crop the image
            cropped_img = image.crop(box)

            # Generate output filename
            filename = os.path.join(output_dir, f"cropped_image1_{i}_{j}.jpg")

            # Save cropped image
            cropped_img.save(filename)


crop_image(image_path, output_directory)
```
